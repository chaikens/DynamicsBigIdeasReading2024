```python
c,z = var('c,z')
```

```python
def f(c, z):
    return z^2 + c
def fn(n, c, z):
    if n == 0:
        return z
    else:
        return f(c,fn(n-1,c,z))
def Fn(n, c, z):
    return( fn(n, c, z) - z )
```

$z$ is periodic of period $n$ for $f(c)$ iff $F_n(n,c,z)=0$.

Lemma: If $m$ divides $n$ then $F_m$ divides $F_n$ in $k[c,z]$.

```python
Phi = [None for a in range(20)]
G   = [None for a in range(20)]
```

```python
Phi[1]=Fn(1, c, z)
G[1] = Phi[1](z=0)
print(Phi[1])
print(G[1])
```

```python
for nn in range(2, 10):
    result = Fn(nn, c, z)
    for div in divisors(nn):
        if div < nn:
            result = result / Phi[div]
    Phi[nn] = factor(result)
    G[nn] = factor(expand(result(z=0)))
    print("\n\n", nn)
    print(Phi[nn])
    print("\n")
    print(G[nn])
```
